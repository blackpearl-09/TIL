### 텍스트 분석 시작, 문자열을 수치 벡터로 만드는 것.
 

- CountVectorizer : 각 텍스트에서 단어 출현 횟수를 카운팅하여 수치벡터화/ 모두 소문자로 변환시킴

  - 출현 횟수를 카운팅하려면 단어사전 필요 -> from sklearn.feature_extraction.text import CountVectorizer
  - 조사, 혹은 지시대명사와 같이 특별한 의미를 지니지 않는데 자주 사용되는 단어는 카운팅에서 높은 횟수를 가지게 되지만, 
실질적으론 의미가 없는 데이터라서 이걸 해결하는 방법으로 나온 것이 바로 TF-IDF   
    
    
    
### TfidfVectorizer : TF-IDF라는 값을 사용하여 CountVectorizer의 단점을 보완함

- TF?
 - Term Frequency의 약자.
 - 1개 문서 안에서 특정 단어의 등장 빈도를 의미
 - 전체 단어수에서 특정 단어가 얼마나 사용되었는지를 파악하여, 해당 문서의 성질을 파악하는 지표로 사용
 - 주로 형태소 단위로 나눔
 
 
- DF?
 - Document Frequency의 약자입니다.
 - 특정 단어가 나타나는 문서 수
 - DF는 한 단어가 전체 문서 집합 내에서 얼마나 공통적으로 많이 등장하는지를 나타내는 값 
 - TF는 주체가 단어라면, DF는 주체가 문서 
 
- IDF? 
 - DF에서 역수 변환을 해준 값
 - 특정 단어가 모든 문서에 다 등장하는 흔한 단어 즉 DF값이 클 수록, tf-idf의 가중치 값을 낮춰주기 위해서 DF값에 역수를 취한 값이 바로 IDF.
 
 
 
 
 ++ 플러스! 특정 단어가 전체의 문서 안에 존재하지 않을 경우,  분모가 0이 되어버리는 상황이 발생. 
  - 이를 방지 하기 위해 분모에 1을 더해줌

```python
#유저 input 받기 (예시, 추후 input으로 받아올 것)

inputs = [first, second, third]

weights = {0:0.5, 1:0.3, 2:0.2}

#for반복문에 index, input <- 두개의 인자를 넣은 이유 
#enumerate: 순서가 있는 자료형(리스트, 튜플, 문자열)을 입력으로 받아 인덱스 값을 포함하는 enumerate 객체를 돌려 줌.


df_temp = df.copy()
df_final = df_temp[['회사명']]
for index, inPut in enumerate(inputs):
    corpus = []
    for value in df[inPut]:  #회사명을 하나씩 value에 넣고
        corpus.append(value)
    vector = CountVectorizer() #countvetorizer 생성
    arr_column = vector.fit_transform(corpus).toarray() #corpus(inPut의 value)로부터 각 단어의 빈도수를 기록
    print(vector.vocabulary_) 
    
    #print(pd.DataFrame(arr_equip).head())
    
    #cosine similarity (기준을 정해서, 유사도를 구해보자)
    #유저가 선택한 기업의 index를 기준으로, 나머지 docs(회사)들의 상대적인 유사도 도출
    
    #그러려면 선택한 회사의 index 번호를 알아야 한다. 그 회사의 행렬을 뽑기 위해서는
    # pandas에 있는, index.values: Return an array representing the data in the Index.
    picked_index = df.index.values[df['회사명']==company][0]
    
    #코사인 유사도 계산함수. !=cosine similarity 공식
    def cos_sim(A,B): 
        np.seterr(divide = 'ignore', invalid = 'ignore') #유효하지 않은 부동 소수점 연산 처리
        return dot(A, B)/(norm(A)*norm(B))
    
# Norm은 벡터의 길이 혹은 크기를 측정하는 방법(함수). 
# Norm이 측정한 벡터의 크기는 원점에서 벡터 좌표까지의 거리 혹은 Magnitude.    
    
    sim_list = []
    
    doc1 = arr_column[picked_index] # 회사명['유저가 선택한 회사의 index번호']
    for doc2 in arr_column : #상대열  
        sim_list.append(cos_sim(doc1,doc2)) #유저가 선택한 회사와 상대 회사들의 코사인유사도 계산
        
    df_temp[inPut] = sim_list #df_temp (만들어 놓은 new데이터프레임) (유저가 입력한 기준 열의) value값을 코사인유사도값으로 대체(위에서 각 inPut의 코사인유사도 구한것)
    df_final[inPut] = [x*weights[index] for x in sim_list]  #위에서 설정한 weight와 index(enumerate로 나온)로 계산, 위의 리스트에서 코사인유사도 값을 하나씩 받아서 
    df_temp.head()
        
df_temp
        
df_final['Total'] = df_final[inputs].sum(axis=1)  


df_final= df_final.sort_values(by='Total', ascending=False)
df_final.head()
# df_final                
# 코사인 유사도 계산 함수. cosine_similarity는 df를 인풋으로 받으나, 이 함수는 numpy array를 input으로 받음.
# 분자가 0일 경우: RuntimeWarning: invalid value encountered in true_divide <- 해당 에러 발생...    
    

```
