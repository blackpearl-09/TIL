## **ACTIVATION FUNCTION**

-- 이전 레이어의 모든 입력에 대한 가중 합을 받아 출력 값을 생성하여 다음 레이어로 전달하는 **비선형 함수**

 -- 딥러닝에서 레이어를 많이 쌓더라도 가중치를 곱하고 편차를 더하는 식이므로 결국 **선형결합**으로 이뤄짐

 -- 선형결합으로 겹겹이 쌓인 layer를 activation function을 통해 non-linearity(비선형성)을 띄도록 하는 이유는, 
 - 데이터가 복잡해지고, feature들의 차원이 증가하면서 데이터 분포가 선형적이지 않고 비선형적으로 나타나면서 
단순한 선형의 boundary로는 표현이 불가능하기 때문에 비선형적인 데이터 분포를 나타내기 위해서 non-linear가 필요

### -- **신호 및 시스템 분야에서 linear와 non-linear에 대한 정의**

 

1. linear의 정의

 - 선형적인 그래프이다.
 - 동차성을 만족한다. (f(ax) = af(x))
 - 가산성을 만족한다. (f(x1 + x2) = f(x1) + f(x2)


2. non-linear는 linear의 성질을 하나라도 만족하지 못하면 non-linear로 정의 (ReLU로 예를들면)
 
  - 1) x = 0에서 그래프가 꺽이기 때문에, 선형적인 그래프가 아니며 
  - 2) f(-3 X 2) ≠ -3f(2) 이고, 
  - 3) f(-3 + 2) ≠ f(-3) + f(2) 이기 때문에 non-linear function 입니다.

### - Vanishing Gradient 
 - 레이어가 깊어질수록 앞선 오차 값이 역방향으로 뒤까지 전달이 되지 않는 문제 발생
 - 미분의 최대값이 0.25인 Sigmoid를 적용 시 역방향으로 0.25보다 작은 값을 곱하고 더해가므로 오차 값이 사그라들어 버림
 - 그 결과, 학습 속도가 크게 저하되거나 학습이 중지됨
 - 이를 보완하기 위해 나온 **ReLU**는 입력값이 양수인 경우에는 gradient가 1로 일정하므로, vanishing gradient를 피할 수 있고 미분도 편리하기 때문에 연산량이 감소된다. ⇒ 식: f(x)=max(0,x)
 - ReLU function의 학습 수렴 속도가 Sigmoid, Tanh보다 6배 정도 빠르다고 알려져 있음
 - ReLU를 이미지 데이터에 많이 사용하는 이유: 이미지는 0~255 사이의 값, RGB로 이뤄져있어 input값이 모두 양수이면서 vanishing gradient를 보완할 수 있는 ReLU를 많이 사용


### - ReLU function의 입력값이 음수인 경우에는 gradient가 무조건 0이 되는데?
 - 이 문제를 극복하기 위해 등장한 activation funciton = Leaky ReLU
 - **Leaky ReLU** : 입력 값이 음수일 때 완만한 선형 함수를 그려준다. 
 - ⇒ 식: f(x)=max(0.01x,x)으로 x가 음수일 때 gradient가 0.01이라는 점을 제외하고는 ReLU와 같다. 
 - ReLU의 발전된 함수 중 **ELU(Exponential Linear Unit)** 도 있다. exponential (지수함수)을 이용해 입력값이 0이하일 때 부드럽게 깎아주기 때문에 미분함수가 곡선형태로 이어져있는 형태다. 

#### ReLU(Rectified Linear Unit)
 - input 되는 x가 0보다 작을 때는 모든 값을 0으로 처리, 0보다 큰 값은 x 그대로 사용.
 - x가 0보다 크기만 하면 미분값이 1이 되므로, 여러 은닉층을 거치며 곱해지더라도 맨 처음 층까지 사라지지 않고 남아있을 수 있다.
 - But, 마지막에 결과 값을 출력하는 Output layer에서는 sigmoid 함수를 사용하게 된다. 0과 1 사이로 나와야 하기 때문에.
 - ReLU의 큰 단점으로 네트워크를 학습할 때 뉴런들이 “죽는”(die) 경우가 발생한다. 
 - ⇒ dying ReLU: 한번 0을 활성화 값으로 다음 레이어에 전달하면, 이후의 뉴런들의 출력값이 모두 0이 되는 현상이 발생. 더이상 값이 업데이트 되지 않게 된다.

#### ELU(Exponential Linear Unit)

 - ReLU의 모든 장점을 포함하며 Dying ReLU 문제를 해결함.
 - Gradient exploding 문제를 방지하진 않는다.
 - 출력 값이 거의 ZERO-CENTERED에 가까우며, 일반적으로 ReLU와 다르게 지수 함수를 계산하는 비용이 발생한다. (Zero-centered? 그래프의 중심 0인 형태로, 함수값이 양수 혹은 음수에만 치우치지 않고 실수 전체에서 나타나는 형태를 의미)
 - exp() 함수를 사용하기 때문에 ReLU 보다는 학습수렴속도가 오래 걸린다.
