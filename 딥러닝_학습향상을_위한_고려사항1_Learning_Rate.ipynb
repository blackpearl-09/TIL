## **딥러닝 학습 향상을 위한 고려 사항들**

1. 학습률 스케쥴러(Learning Rate Scheduler, Lr Scheduler): 미리 학습 일정을 정해두고, 그 일정에 따라 학습률을 조정하는 방법. Pytorch에서는 기본적으로 다양한 learning rate scheduler를 제공한다.  

- 학습률: gradient의 보폭으로, 얼마나 큰 보폭으로 움직일지 결정해주는 값
 ( 하단의 loss 그래프에서 learning rate 설정에 따라 어떻게 최적의 weight을 찾아나가는 지 보여준다.)

2. 학습률 스케쥴러 사용하는 방법 
- optimizer와 scheduler를 먼저 정의한 후, 학습할 때 optimizer.step() 하고 scheduler.step()을 해주면 된다.



3. 학습률 스케쥴러 종류 (지정한 스케쥴러에 따라, pytorch에서의 사용방법이 조금씩 달라짐) 

- 1) LambdaLR: 가장 유연한 learning rate scheduler. 어떻게 scheduling을 할 지 lambda 함수를 이용하여 정해지는데, 설정한 epoch을 lambda 함수에 넣어 나온 값을 초기 learning rate에 곱해서 계산한다. Epoch에 따른 가중치로 lr 를 점점 감소 시키는 방식.   


 image.png  
  
- 2) MultiplicativeLR: 이전 Epoch에 따른 가중치로 lr 를 감소시키는 Scheduler 
 

- 3) StepLR: 가장 흔히 사용되는 learning rate scheduler 중 하나. step size를 설정하여 일정한 Step 마다 learning rate에 gamma를 곱해주는 방식이다. (gamma 비율로 lr를 감소시킨다고 생각)

    < Parameters >
   - optimizer: 이전에 정의한 optimizer 변수명을 넣어준다.
   - step_size: 몇 epoch마다 lr을 감소시킬지가 step_size를 의미한다.
   - gamma: gamma 비율로 lr을 감소시킨다.


- 4) MultiStepLR: StepLR이 균일한 step size를 사용한다면, MultiStepLR은 step size를 여러 기준으로 적용할 수 있는 StepLR의 확장 버전. 여기서는 step_size 대신, milestones에 리스트 형태로 step 범위를 받아 lr를 감소시키는 방식.

    < Parameters >
   - milestones: learning rate 줄일 epoch index의 list
   - gamma: gamma 비율로 lr을 감소시킨다.
