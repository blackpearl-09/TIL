###**Activation Functions**

우리 뇌에 있는 뉴런 기반 모델과 비교할 때 활성화 함수는 다음뉴런으로 무엇을 보낼지 결정하는 마지막에 위치해 있다. 이전 셀의 출력 신호를 받아 다음 셀의 input으로 사용할 수 있게 형식을 변환하는 역할을 한다. 

image.png

- 보다 쉽게 설명하자면, 뉴런은 여러 경로를 통해 들어온 전기신호의 합이 일정치 이상이 되면 다음 뉴런으로 신호를 전달하게 되는 구조이다. 여기서 "일정치 이상이 되는 기준"이 Activation function이 되는 것이다, 예를 들어, 일정 값 이상이 되면 1을 출력하여 다음 뉴런에 값을 전달해주고, 일정 값 이하가 되면 0을 출력하여 다음 뉴런에 값을 전달해주지 않게 된다.



- input 되는 활성화함수는 W*x + b 로, x에 가중치를 곱하고 bias를 더한 값이다. 이렇게 활성화함수를 통해 다음 뉴런으로 들어가는 값을 변환하지 않으면 수백만 개의 파라미터(매개변수)를 가진 심층 신경망의 경우 스케일이 매우 커질 수 있다.
