### 학습 파라미터 초기화 (Weight initialization)

- 배경: Neural Network model의 목적은 손실(loss)을 최소화하는 과정 == 파라미터 최적화(Parameter optimization)
  - 여러 최적화방법 중, loss function을 파라미터 w에 대해 편미분해서 기울기를 계산하는 방식의 경사하강법(Gradient descent)사용하면, 아래의 그래프에서 같은 방식(경사하강법)을 따라 내려가더라도 도달하는 최저점이 다를 수 있다.
  
  

- 목적: 모델을 training하기 전에 신경망 모델의 파라미터에 대한 초기값을 정의하는 데에 사용. 


- 퍼셉트론의 선형결합 결과 값(Activation function의 input값)이 너무 커지거나 작아지지 않도록 하기 위함.
  
  
- **좋은 학습을 위한 조건 중 하나는, 학습 시작 시점의 가중치를 잘 정해주는 것**
  - 이를 위해 파라미터(가중치) 초기화(Weight initialization)방법을 사용

### 1. Xavier Initialization

- 파라미터에 다른 값을 부여하기 위해서 확률분포를 사용하는데, Xavier 초기화는 표준정규분포에 고정된 표준편차를 사용하지 않고 이전 hidden layer의 노드 수에 맞추어 변화시키는 방법. 

- 이전 레이어의 노드의 개수가 n 개이고, 현재 레이어의 노드가 m 개일 때, √ 2 /n+m 을 표준편차로 하는 정규분포로 가중치를 초기화한다. 

- Xaiver 초기화는 비선형함수(ex. sigmoid, tanh)에서 효과적인 결과를 보여준다. 하지만 ReLU함수에서 사용 시 출력 값이 0으로 수렴하게 되는 현상을 확인 할 수 있다. 따라서 ReLU함수에는 또 다른 초기화 방법을 사용해야 한다.
