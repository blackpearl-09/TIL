* 확률적 경사하강법(Stochastic Gradient Descent) 
  - (사전 개념) 경사하강법(Gradient Descent) : 매개변수의 최적값을 찾아내기 위해 손실함수를 미분하여 가장 낮은 경사를 따라 내려가는 방식의 Optimizer. 

  - (배경) GD는 손실함수를 미분하여 기울기를 구하기 위해 매번 모든 트레이닝 데이터를 넣는 방식으로 진행된다. 보다 정확하게 최적값을 찾을수 있으나, 현실적으로 트레이닝 데이터가 수십 만~ 수천 만까지 매우 용량이 많기에 기울기를 구할 때까지 모두 넣어서 계산하면 오랜 시간이 걸린다는 단점이 있다.  
 SGD,GD비교.PNG

  1) 개념 : GD가 가진 이러한 문제를 해결하기 위해 트레이닝 데이터를 mini-batch 단위 만큼 무작위로 추출하여 이 데이터만으로 손실함수를 미분해 기울기를 구하는 학습을 하는 것이 바로 Stochastic Gradient Descent이다. 
  - 무작위, 표본집단을 계속 추출하여 최적값을 구하기 때문에 통계학적으로 GD와 별 차이가 없으나, 시간을 매우 단축할 수 있다는 장점으로 사용되기 시작. 
  - GD가 전체 데치터셋으로 매걸음 방향(매개변수 갱신)을 산정하는 대신 SGD는 Mini Batch를 이용해서 데이터를 조금씩 훑고 쓰윽 지나가며 업데이트한다.
